```python
import numpy as np

def saturation_pressure(Tc: float) -> float:
    return 0.61078 * np.exp((17.27 * Tc) / (Tc + 237.3))

def mixing_ratio(Tc: float, RH: float) -> float:
    p = 101.3
    ph_sat = saturation_pressure(Tc)
    ph = (RH / 100.0) * ph_sat
    return 0.622 * ph / (p - ph)

def coating_thickness_m(w: float, rho: float = 920.0) -> float:
    w_kg_m2 = w / 1000.0
    return w_kg_m2 / rho

def wvtr_physics(Tc: float, RH: float, w: float) -> float:
    omega = mixing_ratio(Tc, RH)
    aT = 1.0 + 0.03 * (Tc - 23.0)
    thickness = coating_thickness_m(w)
    K = 500.0
    WVTR = (K * omega * aT) / thickness
    noise = np.random.normal(0.0, 0.05 * WVTR)
    return WVTR + noise
```

```python
import pandas as pd
import numpy as np

N = 3000

T_list = np.random.uniform(20, 60, N)
RH_list = np.random.uniform(40, 95, N)
w_list = np.random.uniform(10, 60, N)

WVTR_list = [wvtr_physics(T, RH, w) for T, RH, w in zip(T_list, RH_list, w_list)]

df = pd.DataFrame({
    "Temperature_C": T_list,
    "RelativeHumidity_%": RH_list,
    "CoatingWeight_gm2": w_list,
    "WVTR_gm2_day": WVTR_list
})

df.head()
```
