```python
import numpy as np
import pandas as pd
import os

N = 3000  


T_list = np.random.uniform(20, 60, N)      # °C
RH_list = np.random.uniform(40, 95, N)     # %
w_list = np.random.uniform(10, 60, N)      # g/m2

WVTR_list = []

for T, RH, w in zip(T_list, RH_list, w_list):
    WVTR_list.append(wvtr_physics(T, RH, w))

df = pd.DataFrame({
    "Temperature_C": T_list,
    "RelativeHumidity_%": RH_list,
    "CoatingWeight_gm2": w_list,
    "WVTR_gm2_day": WVTR_list
})

os.makedirs("../data", exist_ok=True)  
csv_path = "../data/synthetic_wvtr.csv"

df.to_csv(csv_path, index=False)
csv_path, df.head()
```
